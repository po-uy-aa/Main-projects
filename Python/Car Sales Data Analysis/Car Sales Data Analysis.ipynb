## Q2. Descriptive statistics

With the cleaned data in Q1, please provide the data summarization as below:

* Q2.1 total number of unique name, unique seller_type, unique transmission, and unique owner
* Q2.2 descriptive statistics, including the total number, mean, std, min and max regarding all km_driven
* Q2.3 descriptive statistics, including mean, std, max, and min of the number of selling_price grouped by different seller_type
* Q2.4 descriptive statistics, including mean, std, max, min of the number of years that sold by each owner type


```python
#Q2.1 total number of unique name, unique seller_type, unique transmission, and unique owner
Unique_name=df_delete2['name'].nunique()
Unique_seller_type=df_delete2['seller_type'].nunique()
Unique_transmission=df_delete2['transmission'].nunique()
Unique_owner=df_delete2['owner'].nunique()
print("Total number of unique names:",Unique_name)
print("Total number of unique seller_type:",Unique_seller_type)
print("Total number of unique transmission:",Unique_transmission)
print("Total number of unique owner:",Unique_owner)
```

```python
#Q2.2 descriptive statistics, including the total number, mean, std, min and max regarding all km_driven
print("Total number of Kilometer Driven is:", df_delete2['km_driven'].count())
print("Mean of Kilometer Driven is:", df_delete2['km_driven'].mean())
print("std of Kilometer Driven is:", df_delete2['km_driven'].std() )
print("Min of Kilometer Driven is:" , df_delete2['km_driven'].min())
print("Max of Kilometer Driven is:", df_delete2['km_driven'].max())
```

```python
#Q2.3 descriptive statistics, including mean, std, max, and min of the number of selling_price grouped by different seller_type
sell_price = df_delete2.groupby('seller_type')['selling_price'].count()
Mean = sell_price.mean()
Std = sell_price.std()
Min = sell_price.min()
Max = sell_price.max()
print("Mean of selling prices by seller type:", Mean)
print("Standard deviation of selling prices by seller type:", Std)
print("Min of selling prices by seller type:", Min)
print("Max of selling prices by seller type:", Max)
```

```python
#Q2.4 Descriptive statistics, including mean, std, max, and min of the number of years sold by each owner type
vehicle_age = df_delete2.groupby('owner')['year'].count()
Mean = vehicle_age.mean()
Std = vehicle_age.std()
Min= vehicle_age.min()
Max = vehicle_age.max()
print("Mean of years sold by owner type:", Mean)
print("Standard deviation of the number of years sold by owner type:", Std)
print("Min of years sold by owner type:", Min)
print("Max of years sold by owner type:", Max)
```

## Q4. Detect and remove outliers

We define car sale records WITHOUT following ANY of the below three rules as outliers:

1. owner to be first, second and third owner only
2. Fuel to be Diesel or Petrol only
3. Seller_type to be dealer or individual only

Please remove the corresponding records in the csv file that is not qualified with the rules. After that, __print the length of the data__.

```python
corrected_owners=['First Owner','Second Owner','Third Owner']
corrected_fuels=['Diesel', 'Petrol']
corrected_seller_type=['Dealer', 'Individual']

New_data=df_delete2[
    (df_delete2['owner'].isin(corrected_owners)) &
    (df_delete2['fuel'].isin(corrected_fuels)) &
    (df_delete2['seller_type'].isin(corrected_seller_type)) 
]
print("Length of the changed data is:" , len(New_data))
```
